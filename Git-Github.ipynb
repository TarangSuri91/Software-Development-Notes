## How to start

Either clone or link your local repository with a remote repository where you plan to push your changes.

```python
# git clone <repository-url>
"""
This command is used to create a copy of an existing Git repository, including all of its branches, commits, and history. When you run git clone <repository URL>,
Git creates a new directory containing a copy of the specified remote repository.
It also automatically sets up a remote named "origin" to point to the original repository, from which you cloned.
"""
```

```python
# Working with personal github
# Create new repository in github
# git remote add origin https://github.com//Subhamsharda/splitmate.git : add project as origin
"""
git init                          # Initialize local repo
git add .                         # Stage files
git commit -m "Initial commit"    # Commit changes
git branch -M main                # renames current branch 'main'
git remote add origin <repo-url>  # Link to GitHub
git push origin main              # Push to GitHub
"""
# git remote add origin <repo-url>. This command is used to add a new remote repository.
# When you run it, you're telling Git to associate the alias "origin" with the specified remote repository URL.
# This is done when you want to link your local repository with a remote repository where you plan to push your changes.
```

## Viewing Git Status, History, and Differences

```python
"""
git status
git log: gives commit history in detail
git diff: difference between the changes in your working directory and the last commit
git diff --staged: difference between the changes that are staged (added to the index) and the last commit
"""
```

## Undoing Changes and Resetting Commits

```python
"""
git restore filename	Discard unstaged changes for filename
git restore .	Discard all unstaged changes
git restore --staged filename	Unstage a file but keep changes

git reset --soft HEAD~1
git reset --hard HEAD~1
"""
```

## Working with Git Branches

```python
"""
git branch new-branch-name
git checkout new-branch-name
git checkout main

git branch -v
git branch -r: remote branches
git branch -d branch_name: delete branch
git branch -D branch_name: force delete branch, if branch is not merged
"""
```
